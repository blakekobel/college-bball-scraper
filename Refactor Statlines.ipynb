```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv("old_statlines.csv")
```

```python
kpval = df[['kp_name','KenPomVal','record_date']]
kpval['Metric'] = 'KenPom'
kpval = kpval.rename(columns={"kp_name": "team_name", "KenPomVal": "Value"})

bpival = df[['espn_name','BPIval','record_date']]
bpival['Metric'] = 'BPI'
bpival = bpival.rename(columns={"espn_name": "team_name", "BPIval": "Value"})

srsval = df[['sr_name','SRS','record_date']]
srsval['Metric'] = 'SRS'
srsval = srsval.rename(columns={"sr_name": "team_name", "SRS": "Value"})

oeffval = df[['sr_name','o_eff','record_date']]
oeffval['Metric'] = 'O_Efficiency'
oeffval = oeffval.rename(columns={"sr_name": "team_name", "o_eff": "Value"})

deffval = df[['sr_name','d_eff','record_date']]
deffval['Metric'] = 'D_Efficiency'
deffval = deffval.rename(columns={"sr_name": "team_name", "d_eff": "Value"})

ppgval = df[['sr_name','ppg','record_date']]
ppgval['Metric'] = 'PPG'
ppgval = ppgval.rename(columns={"sr_name": "team_name", "ppg": "Value"})

papgval = df[['sr_name','papg','record_date']]
papgval['Metric'] = 'PAPG'
papgval = papgval.rename(columns={"sr_name": "team_name", "papg": "Value"})

paceval = df[['sr_name','Pace','record_date']]
paceval['Metric'] = 'Pace'
paceval = paceval.rename(columns={"sr_name": "team_name", "Pace": "Value"})

threeparval = df[['sr_name','Three_Par','record_date']]
threeparval['Metric'] = 'Three_Point_Attempt_Rate'
threeparval = threeparval.rename(columns={"sr_name": "team_name", "Three_Par": "Value"})

trueshootval = df[['sr_name','True_Shoot','record_date']]
trueshootval['Metric'] = 'True_Shooting'
trueshootval = trueshootval.rename(columns={"sr_name": "team_name", "True_Shoot": "Value"})

tovval = df[['sr_name','Tov','record_date']]
tovval['Metric'] = 'Turnover_Perc'
tovval = tovval.rename(columns={"sr_name": "team_name", "Tov": "Value"})

oppthreeparval = df[['sr_name','Opp_Three_Par','record_date']]
oppthreeparval['Metric'] = 'Opp_Three_Point_Attempt_Rate'
oppthreeparval = oppthreeparval.rename(columns={"sr_name": "team_name", "Opp_Three_Par": "Value"})

opptrueshootval = df[['sr_name','Opp_True_Shoot','record_date']]
opptrueshootval['Metric'] = 'Opp_True_Shooting'
opptrueshootval = opptrueshootval.rename(columns={"sr_name": "team_name", "Opp_True_Shoot": "Value"})

opptovval = df[['sr_name','Opp_Tov','record_date']]
opptovval['Metric'] = 'Opp_Turnover_Perc'
opptovval = opptovval.rename(columns={"sr_name": "team_name", "Opp_Tov": "Value"})
```

```python
result = kpval.append([bpival,srsval,oeffval,deffval,ppgval,papgval,paceval,threeparval,trueshootval,tovval,oppthreeparval,opptrueshootval,opptovval])
```

```python
result = result[['team_name','Metric','Value','record_date']]
```
