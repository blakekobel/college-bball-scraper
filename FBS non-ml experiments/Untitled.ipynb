```python
import json
import requests
import datetime as dt
import pickle
import pandas as pd
import mysql.connector
import os
from os.path import join, dirname
from datetime import date, datetime
from dotenv import load_dotenv
# import db_helper
```

```python
sports_response = requests.get('http://localhost:8000/suggestedBets/current')
file = json.loads(sports_response.text)
```

```python
file
```

```python
sorted_Dict = sorted(file, key=lambda x: x['suggestedBet']['start_time'])
for game in sorted_Dict:
    print(game['suggestedBet']['start_time'])
    tipOff = game['suggestedBet']['start_time'].split("T")[1]
    tipOff = dt.datetime.strptime(tipOff,'%H:%M:%S').strftime('%I:%M %p')
    print(tipOff)
```

```python
all_game_html = ''
for game in file:
    away = game['awayTeam']['sr_name']        
    home = game['homeTeam']['sr_name']
    pick = game['pick']['sr_name']
    if pick == home:
        opponent = away
    else:
        opponent = home
    site = game['suggestedBet']['bet_site']
    odds = game['suggestedBet']['odds']
    minOdds = game['suggestedBet']['minimum_suggested_odds']
    tipOff = game['suggestedBet']['start_time'].split("T")[1]
    tipOff = dt.datetime.strptime(tipOff,'%H:%M:%S').strftime('%I:%M %p')
    
    html_string = f'<p style="font-size:16px;line-height:1.5em;font-family:Georgia,Times,serif;margin-bottom:1em">{tipOff} : {pick} needs {minOdds} over {opponent}. {site} has {odds} odds</p><hr style="font-family:Georgia,Times,serif">'
    all_game_html += html_string
print(all_game_html)
```
